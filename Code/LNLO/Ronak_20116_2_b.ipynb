```python
from sympy import *
x= symbols('x')
u = symbols('u',nonnegative=True) # u > 0
```

```python
from sympy import lambdify
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.arange(-100,101,1)
x = symbols('x')
y = x**2 + 1
f = lambdify(x,y,"numpy")
```

```python
plt.plot(a,f(a))
plt.xlabel('x')
plt.ylabel('x^2 + 1')
plt.title("Objective Function")
plt.show()
```

```python
u = symbols('u')
expr = y + u * (x**2 - 6*x + 8)
L = lambdify([x,u],expr,"numpy")
```

```python
plt.plot(a,L(a,0.01),label="u = 0.01")
plt.plot(a,L(a,0.1),label="u = 0.1")
plt.plot(a,L(a,1),label="u = 1")
plt.plot(a,L(a,10),label="u = 10")
plt.legend()
plt.xlabel('x')
plt.ylabel('L(x,u)')
plt.title('L(x,u) vs x')
plt.show()
```

```python
expr1 = (9*u + 1 - u**2)/(1+u)
g = lambdify(u,expr1,"numpy")
a1 = np.array([0.01,0.1,1,10])
plt.plot(np.log10(a1),g(a1))
plt.xlabel('log10(u)')
plt.ylabel('g(u)')
plt.title('g(u) vs log(u)')
```
